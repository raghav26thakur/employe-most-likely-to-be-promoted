```python
# Splitting the dataset into training and testing using 70% and 30% respectively

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.3)
```

```python
# Import the Gaussian Naive Nayes classifier

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# Implementing the Gaussian Naive Bayes model

model.fit(x_train, y_train)
```

```python
# Determining the accuracy score of the implemented Naive Bayes model

model.score(x_test, y_test)
```
